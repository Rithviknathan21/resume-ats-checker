```python
import re
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    return text

def analyze_resume(resume, job_desc):
    resume_clean = preprocess(resume)
    job_clean = preprocess(job_desc)

    vectorizer = CountVectorizer().fit_transform([resume_clean, job_clean])
    vectors = vectorizer.toarray()

    similarity = cosine_similarity([vectors[0]], [vectors[1]])[0][0]

    resume_words = set(resume_clean.split())
    job_words = set(job_clean.split())

    missing_keywords = job_words - resume_words

    result = f"""
Match Score: {round(similarity * 100, 2)} %

Missing Keywords (Top 10):
{list(missing_keywords)[:10]}

Suggestion:
"""

    if similarity > 0.7:
        result += "Good match! Just refine wording."
    elif similarity > 0.4:
        result += "Moderate match. Add missing skills."
    else:
        result += "Low match. Improve resume with job keywords."

    return result
```

```python
resume_text = input("Paste your RESUME text:\n")
job_description = input("\nPaste JOB DESCRIPTION:\n")

output = analyze_resume(resume_text, job_description)

print(output)

# Save result to Drive
file_path = "/content/drive/MyDrive/Smart Resume Keyword Analyzer/result.txt"

with open(file_path, "w") as f:
    f.write(output)

print("\n✅ Result saved to:", file_path)
```
